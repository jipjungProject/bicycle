```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error 
from sklearn.model_selection import GridSearchCV 
from sklearn import linear_model as lm 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import seaborn as sns
import xgboost as xgb 
from scipy.stats import uniform, randint
import numpy as np

n = 3
results = {} 
x_test1= pd.DataFrame()
y_test1= pd.DataFrame()

total = df1['대여장소'].nunique()  # 서로다른 대기소 갯수 10개
# print(total)  # 10
a=0

for idx, (name, _df) in enumerate(df1.groupby('대여장소'), 1):
    print()
    print("[%d/%d] %d번 대여소" %(idx, total, name), end=' ')  #name은 대여소 코드.
    print()
    
    try:
        _df = _df.groupby(col_group)['이용건수'].sum().reset_index() # 이용건수 취합
        _df = _df.join(pd.get_dummies(_df['요일'], prefix="요일")) # 요일을 one-hot-encoding
#         print(_df)
        
        # 이값이 출력되는 ()쌍. 
        # (대여소에 1년기준 몇행이 있는지 * 컬럼의 수). 1년 전체에서 각 타임별로 나눠서 취합한 데이터의 행
        print(_df.shape)

        # 한 대여소의 1년 데이터를 split한다. 레이블로 _df['이용건수']가 사용됨        
        x_train, x_test, y_train, y_test = train_test_split(_df[feature], _df['이용건수'], test_size=0.2, random_state=5) 
        if(a==0):
            x_test1=x_test
            y_test1=y_test
            a=a+1


        # 그리드 서치를 이용해서 모델링
        param_grid = { 
            "gamma": uniform(0, 0.5).rvs(n), 
            "max_depth": range(2, 7), # default 3 
            "n_estimators": randint(100, 150).rvs(n), # default 100 
        } 
        
        
        grid = GridSearchCV(xgb.XGBRegressor(objective='reg:squarederror'), cv=5, n_jobs=-1, param_grid=param_grid, verbose=1) 
        grid = grid.fit(x_train, y_train, verbose=1) 
        
        
        model = grid.best_estimator_ 
        predictions = model.predict(x_test) 
        mae = mean_absolute_error(y_test, predictions) 
        
        
        # 모델저장
        
#         model.save_model("my_model.model")

        model.save_model("data/model/my_model" + str(int(name)) + ".model")
        
        # y_test는 pandas.series 형식이다. index와 value로 이루어지므로,
        # 값을 참조하려면 y_test.values[행번호]로 한다.
        
#         for i in range(0, 5):
#             print('>predicted='+ str(predictions[i]) +', actual=' + str(y_test.values[i]))
        

        # results[대여소코드]에 결과 저장
        results[name] = {} 
        results[name]['model'] = model  #best_estimator 저장
        results[name]['mae'] = mae 
        results[name]['errors'] = predictions - y_test  
    
    except:
        continue
    
    
errors = np.array([result['mae'] for result in results.values()]) 
errors.mean()







```
